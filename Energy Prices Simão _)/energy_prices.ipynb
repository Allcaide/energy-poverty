```python
energy_prices_df2.columns = [
    "Banda",
    "Consumo_Min",
    "Consumo_Max",
    "Preco_Sem_Impostos",
    "Preco_Sem_IVA",
    "Preco_Final"
]

print(energy_prices_df2)
```

```python
#Toremove the lines that contain just titles or blank lines
clean_df = energy_prices_df2[energy_prices_df2["Banda"].astype(str).str.contains("Banda -")]
clean_df.head(10)
```

```python
year = 2025
years = []
for i in range(len(clean_df)):
    years.append(year)
    if(i + 1)%10== 0:
        year -= 1

clean_df["Ano"]= years
```

```python
clean_df.head(20)
```

```python
clean_df = clean_df.drop(columns = ["Preco_Sem_IVA", "Preco_Sem_Impostos"])
clean_df
```

```python
clean_df["Consumo_Max"] = clean_df["Consumo_Max"].fillna("Sem limite")
```

```python
clean_df = clean_df.sort_values("Ano", ascending = False)
clean_df
```

```python
clean_df
```

```python
clean_df_da = (
    clean_df
    .reset_index()
    .query("Banda == 'Banda - DC'")#referenciar estatisticas da erce
)
```

```python
clean_df_da = clean_df_da.drop(columns = ['Consumo_Min','Consumo_Max'])
```
