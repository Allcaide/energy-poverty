We are going to eliminate values equal to "Serviços" in column "Tipo de Edifício". Energy poverty has to do with habitational buildings only.

```python
dfclean=df2015_2025.drop(df2015_2025[df2015_2025['Tipo de Edifício'] == 'Serviços'].index)
```

```python
dfclean.shape
```

```python
dfclean['Tipo de Edifício'].unique()
```

```python
dfclean['Classe Energética'].unique()
```

We dropped coluns "Total de Emissões CO2" and "Área Média Útil de Pavimento", because those are not useful for energy poverty assessment.

```python
dfclean=dfclean.drop(columns=['Total de Emissões CO2','Área Média Útil de Pavimento'])
```

```python
dfclean.columns
```

```python
dfclean['Tipo de Doc / Contexto'].value_counts()
```

A big part of the buildings are existing buildings, which means that, in our dataframe, we have certificates of new buildings and existing buildings.
We deleted the rows about the certificate of buildings still at the project stage because these buildings do not exist yet.

```python
dfclean=dfclean.drop(dfclean[dfclean['Tipo de Doc / Contexto'] == 'Projeto Edifícios Novos'].index)
dfclean=dfclean.drop(dfclean[dfclean['Tipo de Doc / Contexto'] == 'Projeto Renovação Edifícios'].index)
```

```python
dfclean['Tipo de Edifício'].value_counts()
```

Now that all building types, are the same, we can erase this column.

```python
dfclean=dfclean.drop(columns=['Tipo de Edifício'])
```

```python
dfclean["Mês de Emissão"].dtype
```

We want the dataframe to be by year and not by month.

```python
dfclean["Mês de Emissão"] = pd.to_datetime(dfclean["Mês de Emissão"])
```

```python
dfclean["Mês de Emissão"] = dfclean["Mês de Emissão"].dt.year
```

```python
df_clean=(dfclean.groupby(['Mês de Emissão', 'Região', 'NUTS III', 'Distrito', 'Concelho', 'Tipo de Doc / Contexto', 'Classe Energética'], as_index=False)['Total de Certificados Emitidos'].sum())
```
